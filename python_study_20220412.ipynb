```python
import time
start = time.time()
folded_thickness = 0.00008
num = 0

while num < 43 :
    folded_thickness *= 2
    num += 1

print("厚さ： {:.2f}キロメートル".format(folded_thickness / 1000))
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
import time
start = time.time()
folded_thickness = 0.00008
for num in range(43):
  folded_thickness *= 2

print("厚さ： {:.2f}キロメートル、回数： {}回".format(folded_thickness / 1000, num+1))
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

```python
%matplotlib inline
import time
import matplotlib.pyplot as plt
start = time.time()

# リスト初期化
l=[]
l.append(0) # 0点登録
folded_thickness = 0.00008 # 紙の厚み初期値

# while num < 43 :
#  folded_thickness *= 2
#  num += 1

for num in range(43):
  folded_thickness *= 2
  l.append(folded_thickness)

print(l)
# グラフを表示する。タイトルと軸ラベル名付き。
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness [km/100000]")
plt.plot(l, color='red')
plt.show()
elapsed_time = time.time() - start
print("厚さ： {:.2f}キロメートル、回数： {}回".format(folded_thickness / 1000, len(l)-1))
print("time : {}[s]".format(elapsed_time))
```
